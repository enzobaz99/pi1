Creamos los "id" de cada plataforma convinando la inicial de cada plataforma con show_id

```python
df_amazon.insert(loc=0, column="id", value="a" + df_amazon["show_id"])
df_disney_plus.insert(loc=0, column="id", value="dp" + df_disney_plus["show_id"])
df_hulu.insert(loc=0, column="id", value="h" + df_hulu["show_id"])
df_netflix.insert(loc=0, column="id", value="n" + df_netflix["show_id"])
```

Unimos todos los dataframes

```python
df_todaslasplataformas = df_amazon.append(df_disney_plus).append(df_hulu).append(df_netflix)
```

cambiamos los valores nulos del campo rating reemplazandoce por el string “ G” (corresponde al rating de madurez: “general para todos los públicos”

```python
df_todaslasplataformas["rating"].fillna("G", inplace=True)
```

Revisamos los cambios

```python
df_todaslasplataformas.head(2)
```

Cambiamos el formato de la fecha (date_added) en AAAA/MM/DD

```python
df_todaslasplataformas["date_added"] = pd.to_datetime(df_todaslasplataformas["date_added"])
```

Revisamos los cambios

```python
df_todaslasplataformas.head(2)
```

Cambiamos las letras mayusculas a minusculas y revisamos si ocurrio

```python
df_todaslasplataformas["type"] = df_todaslasplataformas["type"].str.lower()
df_todaslasplataformas["title"] = df_todaslasplataformas["title"].str.lower()
df_todaslasplataformas["director"] = df_todaslasplataformas["director"].str.lower()
df_todaslasplataformas["cast"] = df_todaslasplataformas["cast"].str.lower()
df_todaslasplataformas["country"] = df_todaslasplataformas["country"].str.lower()
df_todaslasplataformas["listed_in"] = df_todaslasplataformas["listed_in"].str.lower()
df_todaslasplataformas["description"] = df_todaslasplataformas["description"].str.lower()
df_todaslasplataformas["rating"] = df_todaslasplataformas["rating"].str.lower()
df_todaslasplataformas["duration"] = df_todaslasplataformas["duration"].str.lower()
df_todaslasplataformas
```

Dividimos la columna "duration" en "duration_type" y "duration_int" separando asi los "min" y "season" de lo que aparece en numeros

```python
divisor = df_todaslasplataformas["duration"].str.split(" ", n=1, expand=True)
df_todaslasplataformas["duration_int"] = divisor[0]
df_todaslasplataformas["duration_type"] = divisor[1]
df_todaslasplataformas["duration_type"] = df_todaslasplataformas["duration_type"].str.replace("seasons", "season")
df_todaslasplataformas["duration_int"] = df_todaslasplataformas["duration_int"].astype("Int64")
df_todaslasplataformas = df_todaslasplataformas.drop('duration', axis=1)
df_todaslasplataformas.head(10)
```

Guardamos el dataframe

```python
df_todaslasplataformas.to_csv("todaslasplataformas_database.csv", index=False)
```
